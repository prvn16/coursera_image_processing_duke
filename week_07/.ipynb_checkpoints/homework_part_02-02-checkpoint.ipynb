## Week 7 Part 2
For a given image, compute at every pixel the inner product between the gradient of the Laplacian and the level lines normal, this being the main term in one of the inpainting techniques we learned. Display it and analyze its behavior. 

```python
from __future__ import division, print_function

import numpy as np
from matplotlib import pyplot as plt
import time

from skimage import img_as_float
from skimage import io
from skimage.color import rgb2gray

%matplotlib inline

# Plotting functions

def plot_1xc(imgs_list, titles_list, save_file=None):
    cols = len(imgs_list)
    i = 0
    
    fig, axes = plt.subplots(nrows=1, ncols=cols, figsize=(15,15))
    for c in range(cols):
        axes[c].imshow(imgs_list[i], cmap="gray", interpolation="nearest")
        axes[c].set_title(titles_list[i], size=20)
        axes[c].set_xticks([])
        axes[c].set_yticks([])
        i = i + 1
    plt.tight_layout();
    
    if not (save_file == None):
        filename = time.strftime("%Y%m%d_%H%M") + "_" + save_file + ".png"
        fig.savefig(filename, bbox_inches='tight')
        
def plot_rxc(imgs_list, titles_list, ncol=3, save_file=None):
    cols = ncol
    rows = np.ceil(len(imgs_list)/cols).astype(int)
    if rows==1:
        plot_1xc(imgs_list, titles_list, save_file)
    else:
        i = 0
    
        fig, axes = plt.subplots(nrows=rows, ncols=cols, figsize=(15,15*rows/cols))
        for r in range(rows):
            for c in range(cols):
                if i < len(imgs_list):
                    axes[r,c].imshow(imgs_list[i], cmap="gray", interpolation="nearest")
                    axes[r,c].set_title(titles_list[i], size=20)
                    axes[r,c].set_xticks([])
                    axes[r,c].set_yticks([])
                i = i + 1
        plt.tight_layout();
    
        if not (save_file == None):
            filename = time.strftime("%Y%m%d_%H%M") + "_" + save_file + ".png"
            fig.savefig(filename, bbox_inches='tight')
```

```python
# import numpy as np
import cv2

img = cv2.imread('../images/donut.png')
mask = cv2.imread('../images/donut_mask.png',0)
dst = cv2.inpaint(img,mask,5,cv2.INPAINT_TELEA)
dst2 = cv2.inpaint(img,mask,5,cv2.INPAINT_NS)

img = img[:,:,-1]
dst = dst[:,:,-1]
dst2 = dst2[:,:,-1]

print(np.min(img), np.max(img), np.min(mask), np.max(mask), np.min(dst), np.max(dst), np.min(dst2), np.max(dst2))

plot_rxc([img, mask, dst, dst2], ["img", "mask", "INPAINT_TELEA", "INPAINT_NS"], ncol=2)
```

```python
import cv2

img = cv2.imread('../images/donut2.png')
mask = cv2.imread('../images/donut2_mask.png',0)
dst = cv2.inpaint(img,mask,5,cv2.INPAINT_TELEA)
dst2 = cv2.inpaint(img,mask,5,cv2.INPAINT_NS)

img = img[:,:,-1]
dst = dst[:,:,-1]
dst2 = dst2[:,:,-1]

dst_detail = dst[50:75, 125:200]
dst2_detail = dst2[50:75, 125:200]


print(np.min(img), np.max(img), np.min(mask), np.max(mask), np.min(dst), np.max(dst), np.min(dst2), np.max(dst2))

ims = [img, mask, dst, dst2, dst_detail, dst2_detail]
titles = ["img", "mask", "INPAINT_TELEA", "INPAINT_NS", "INPAINT_TELEA Detail", "INPAINT_NS Detail"]
plot_rxc(ims, titles, ncol=2)
```
