```python
from __future__ import division, print_function

import numpy as np
from matplotlib import pyplot as plt
import time
import sys
import Queue

import skimage
from skimage import img_as_float, img_as_ubyte
from skimage import io
from skimage.color import rgb2gray
from skimage.util import random_noise
from skimage import transform

%matplotlib inline
```

```python
def plot_1xc(imgs_list, titles_list, save_file=None):
    cols = len(imgs_list)
    i = 0
    
    fig, axes = plt.subplots(nrows=1, ncols=cols, figsize=(15,15))
    for c in range(cols):
        axes[c].imshow(imgs_list[i], cmap="gray")
        axes[c].set_title(titles_list[i], size=20)
        axes[c].set_xticks([])
        axes[c].set_yticks([])
        i = i + 1
    plt.tight_layout();
    
    if not (save_file == None):
        filename = time.strftime("%Y%m%d_%H%M") + "_" + save_file + ".png"
        fig.savefig(filename, bbox_inches='tight')
```

```python
def check_valid_loc(img, loc):
    """Check if 'loc' is a valid location in 'img'"""
    xmin = 0; xmax=img.shape[0] - 1
    ymin = 0; ymax=img.shape[1] - 1
    x = loc[0]
    y = loc[1]
    if (x >= xmin) & (x <= xmax) & (y >= ymin) & (y <= ymax):
        return True
    else:
        return False
```

```python
def get_neighbor_locs(img, loc, size):
    """Return valid 'size'-neighborhood location tuples of 'loc' pixel"""
    locs = []
    offset = np.floor(size/2.0).astype(int)
    for dx in range(-offset,offset+1):
        for dy in range(-offset,offset+1):
            if not ((dx == 0) & (dy == 0)): # center pixel is not a neighbor of itself
                possible = (loc[0] + dx, loc[1] + dy)
                if check_valid_loc(img, possible):
                    locs.append(possible)
    return locs            
```

```python
# img = rgb2gray(io.imread("../images/circle_100.png"))
# img = rgb2gray(io.imread("../images/circle.png"))
# img = rgb2gray(io.imread("../images/twister_simplified.jpg"))
# img = rgb2gray(io.imread("../images/twister.jpg"))
img = rgb2gray(io.imread("../images/synapse.jpeg"))
# img = rgb2gray(io.imread("../images/maze.jpg"))

img = img_as_float(img)
seed = (85,85)
# seed = (3,3)
seed_value = img[seed]
# print(seed_value, img.shape)
value_tolerance = 0.35

UNEXAMINED = 0
QUEUED = 1
OUT = 2
IN = 3

status = np.zeros_like(img).astype(int) # Keeps track of pixel status
status.fill(UNEXAMINED)

q = Queue.Queue()
q.put(seed)

while not q.empty():
    loc = q.get()
    
    if np.abs(seed_value - img[loc]) <= value_tolerance:
        status[loc] = IN
    else:
        status[loc] = OUT
    
    if status[loc] != OUT:
        neighbor_locs = get_neighbor_locs(img, loc, 21)
        for nl in neighbor_locs:
            if status[nl] == UNEXAMINED:
                status[nl] = QUEUED
                q.put(nl)

mask = status == IN
mask2 = status != IN
img2 = img.copy()
img2[mask] = 0
img2[mask2] = 1

titles = ["Original", "Segmented, seed=" + str(seed) + ", thresh=" + str(value_tolerance)]
plot_1xc([img, img2], titles, save_file="part5_segmented")
```
